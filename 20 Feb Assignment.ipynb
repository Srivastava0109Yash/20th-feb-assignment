## Q1. Explain GET and POST methods.

__GET method__:
GET is a request method used to retrieve data from the server.
It sends data in the URL query string, which is visible in the browser's address bar.
GET requests are idempotent, meaning that sending the same request multiple times will always result in the same response.
GET requests are cached by the browser, so subsequent requests for the same resource may be served from the cache, improving performance.

__POST method__:
POST is a request method used to submit data to be processed by the server.
It sends data in the request body, which is not visible in the browser's address bar.
POST requests are not idempotent, meaning that sending the same request multiple times may result in different responses.
POST requests are not cached by the browser, so subsequent requests for the same resource will always be sent to the server.

## Q2. Why is request used in Flask?

In Flask, request is a built-in object that represents an incoming HTTP request from a client. It provides access to the data that the client sends to the server, such as form data, query parameters, and HTTP headers.

The request object can be used to:

__Retrieve form data__: Flask allows developers to receive form data submitted from HTML forms using the request.form attribute. This attribute provides a dictionary-like interface for accessing form data.

__Access query parameters__: Query parameters are used to provide additional information to the server, and are appended to the URL of a request. Flask provides a request.args attribute to access the query parameters.

__Retrieve JSON data__: JSON is a popular data format for APIs. Flask can parse incoming JSON data using the request.get_json() method.

__Access cookies__: Cookies are small text files that are stored on a user's computer. Flask provides a request.cookies attribute to access cookies sent by the client.

__Access HTTP headers__: HTTP headers contain additional information about the request, such as the User-Agent, Content-Type, and Accept headers. Flask provides a request.headers attribute to access these headers.

## Q3. Why is redirect() used in Flask?

The redirect()  function in Flask is used to redirect a user to a different URL. It is typically used when a user submits a form, and the server needs to send the user to a different page, such as a confirmation page or a login page.

Here are some common use cases for the redirect() function in Flask:

__After a successful form submission__:
When a user submits a form, the server may need to redirect the user to a different page to display a confirmation message or a summary of the submitted data.

__After a login or logout action__: 
When a user logs in or logs out of a website, the server may need to redirect the user to a different page, such as the home page or the login page.

__When a URL has changed__: 
If a URL has changed, the server can use the redirect() function to automatically send users to the new URL.

## Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are files that contain the HTML code and placeholders for dynamic content. They are used to create dynamic web pages that display data from the server or user input. Flask uses a templating engine, called Jinja2, to render templates.

The render_template() function in Flask is used to render a template and return the resulting HTML to the client. It takes the name of the template file as an argument, along with any data that needs to be passed to the template.
